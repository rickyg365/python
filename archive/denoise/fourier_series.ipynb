## Fourier Transform and Plot Frequency Spectrum Notes

Source: [How to Compute FFT and Plot Frequency Spectrum...](https://www.youtube.com/watch?v=O0Y8FChBaFU) by [1M views](https://www.youtube.com/channel/UC8DvpZf4U_vGN0YGxh77edQ)

```python
import numpy as np
import matplotlib.pyplot as plt

# Construct Time Signal
Fs = 2000  # Sampling freq, look up sampling theory
dt = 1 / Fs  # Sample Time Interval
f0 = 100 # Signal freq

N = int(10 * Fs / f0)  # number of samples, increase for finer resolution in freq domain

t = np.linspace(0, (N-1)*dt, N)  # Time steps
fstep = Fs / N  #  freq interval

f = np.linspace(0, (N-1)*fstep, N)  # freq steps

y = 1 * np.sin(2*np.pi * f0 * t) + 4 * np.sin(2*np.pi * 2*f0 * t)  # Magnitude and freq
# Add constant for DC

# FFT
x = np.fft.fft(y)  # Series of complex nums
x_mag = np.abs(x) / N  # Normalize by num of samples

f_plot = f[0:int(N/2+1)]
x_mag_plot = 2 * x_mag[0:int(N/2+1)]
x_mag_plot[0] = x_mag_plot[0] / 2 # DC component does not need to multiply


# Plot
fig, [ax1, ax2] = plt.subplots(2,1)
ax1.plot(t,y, '.-')
ax2.plot(f_plot, x_mag_plot, '.-')


ax1.set_xlabel('time (s)')
ax2.set_xlabel('freq (Hz)')

ax1.grid()
ax2.grid()

ax1.set_xlim(t[0], t[-1])
ax2.set_xlim(f_plot[0], f_plot[-1])

plt.tight_layout()
```
